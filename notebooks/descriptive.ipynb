```python
import pandas as pd
#pip install dash
import dash
from dash import dcc, html
import plotly.express as px

# Load and prepare data
df = pd.read_csv('../data/processed/final_dataset.csv')
df = df.set_index(['code', 'year']).reset_index()

# Select variables of interest
variables = {
    'Digital Payment': 'Made or received a digital payment, rural (% age 15+)',
    'Food Stability': 'foodSupply_stability',
    'Production Value': 'Agriculture, forestry, and fishing, value added (percent of GDP)',
    'Crop Production Index': 'Crop production index (2014-2016 = 100)'
}

# Create list of countries from the dataset
country_codes = df['code'].unique()

# Initialize Dash app
app = dash.Dash(__name__)
app.title = "Country Statistics Dashboard"

# App layout
app.layout = html.Div([
    html.H1("Descriptive Statistics Dashboard", style={'textAlign': 'center'}),
    
    html.Label("Select a Country:"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': code, 'value': code} for code in country_codes],
        value=country_codes[0]
    ),
    
    html.Div(id='country-summary')
])

# Callback to update dashboard
@app.callback(
    dash.dependencies.Output('country-summary', 'children'),
    [dash.dependencies.Input('country-dropdown', 'value')]
)
def update_dashboard(selected_code):
    df_country = df[df['code'] == selected_code]
    summary = df_country[variables.values()].describe().T.round(2)
    
    fig = px.line(df_country, x='year', y=list(variables.values()), 
                  title=f"Time Trends in {selected_code}")
    
    return html.Div([
        html.H3(f"Descriptive Stats for {selected_code}"),
        html.Table([
            html.Thead([
                html.Tr([html.Th(col) for col in ['Variable', 'Count', 'Mean', 'Std', 'Min', '25%', '50%', '75%', 'Max']])
            ]),
            html.Tbody([
                html.Tr([html.Td(index)] + [html.Td(summary.loc[index, col]) for col in summary.columns])
                for index in summary.index
            ])
        ], style={'width': '100%', 'border': '1px solid gray', 'borderCollapse': 'collapse'}),
        dcc.Graph(figure=fig)
    ])

# Run app
if __name__ == '__main__':
    app.run(debug=True)
```
